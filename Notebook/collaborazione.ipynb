## CREAZIONE RELAZIONE COLLABORAZIONE

```python
pivot_autori['collaboratore'] = pivot_autori['Autori']
```

```python
collab = pivot_autori.groupby(['id_track']).agg({'Autori':list,'collaboratore':list}).reset_index()
collab = collab.explode('Autori')
collab = collab.explode('collaboratore').reset_index(drop=True)
collab
```

```python
same_person = []
for i, x in enumerate (collab.iterrows()):
    autore = collab.loc[i,'Autori']
    collaboratore = collab.loc[i,'collaboratore']
    if autore == collaboratore:
        same_person.append('stesso')
    else:
        same_person.append('diverso')
collab['persona'] = same_person      
collab.head(5)
```

```python
## primo filtro
collab = collab[collab['persona']!='stesso'].reset_index(drop=True)
collab.head(5)
```

```python
collab['inverso']='x'
for i, x in enumerate(collab.iterrows()):
    for a, b in enumerate(collab.id_track):
        if collab.loc[i, 'id_track'] == b and i!=a:
          if collab.loc[i, 'Autori'] == collab.loc[a, 'collaboratore'] and collab.loc[i, 'collaboratore'] == collab.loc[a, 'Autori']:
            if i > a:
              collab.loc[i, 'inverso']='non va bene'
            else:
              collab.loc[i, 'inverso']='va bene'              
collab.head(5)
```

```python
collab = collab[collab['inverso']!='non va bene'].reset_index(drop=True)
print(collab.shape[0])
collab.head(5)
```

```python
## merge per recuperare edizione
collab = collab.merge(tracce[['id_track', 'edizione', 'titolo']], on='id_track', how='left')
print(collab.shape[0])
collab.head(5)
```

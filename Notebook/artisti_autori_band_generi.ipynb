## autori

```python
pivot_autori = pd.read_csv('/content/drive/MyDrive/Data_man_Project/Data_Man_project/Data/pivot_autori_final.csv')
print(pivot_autori.shape)
print('autori_unici:',pivot_autori.groupby('Autori').count().shape[0])
print('track_unici:',pivot_autori.groupby('id_track').count().shape[0])
pivot_autori.head(2)
```

## creazione solo_autori, artisti_autori, solo_artisti

```python
solo_cantanti = pd.read_csv('/content/drive/MyDrive/Data_man_Project/Data_Man_project/Data/SOLO_CANTANTI_artisti_final.csv')
print(solo_cantanti.shape)
print('solo_cantanti:',solo_cantanti.groupby('id_artist').count().shape[0])
solo_cantanti.nome = solo_cantanti.nome.apply(lambda x: str.lower(x))
solo_cantanti.head(5)
```

```python
solo_cantanti.isna().sum()
```

```python
## separare autori e artisti_autori
autori2 = solo_cantanti.merge(pivot_autori, left_on='nome', right_on='Autori', how='right')
print(autori2.shape)
print('autori_unici:',autori2.groupby('Autori').count().shape[0])
print('track_unici:',autori2.groupby('id_track').count().shape[0])
autori2.head(3)
```

```python
autori2 = autori2.fillna('0')
autori2.head(3)
```

```python
## quelli con id_artist !=0 significa che sono sia autori che artisti, quindi artisti_autori
artisti_autori = autori2[autori2['id_artist']!='0'].reset_index(drop=True)
print(artisti_autori.shape)
print('artisti_autori:',artisti_autori.groupby('id_artist').count().shape[0])
print('track_unici:',artisti_autori.groupby('id_track').count().shape[0])
artisti_autori.drop(columns='edizione', inplace=True)
artisti_autori = artisti_autori.drop_duplicates().reset_index(drop=True)
print('POST RIMOZIONE MISSING VALUES')
print(artisti_autori.shape)
print('artisti_autori:',artisti_autori.groupby('id_artist').count().shape[0])
print('track_unici:',artisti_autori.groupby('id_track').count().shape[0])
artisti_autori.head(3)
```

```python
## quelli con id_artist == 0 significa che sono solo_autori
solo_autori = autori2[autori2['id_artist']=='0'].reset_index(drop=True)
print(solo_autori.shape)
print('solo_autori:', solo_autori.groupby('Autori').count().shape[0])
print('track_unici:', solo_autori.groupby('id_track').count().shape[0])
solo_autori.head(3)
```

```python
print('num_tot_autori:', autori2.groupby('Autori').count().shape[0])
print('num_solo_autori:', solo_autori.groupby('Autori').count().shape[0])
print('num_artisti_autori:', artisti_autori.groupby('id_artist').count().shape[0])
```

```python
solo_artisti= artisti_autori.groupby('id_artist').agg({'nome':'first'}).reset_index()
solo_artisti = solo_artisti.merge(solo_cantanti, on=['id_artist'], how='right')
solo_artisti.fillna(0, inplace=True)
print(solo_artisti.shape)
print('solo_artisti:',solo_artisti.groupby('id_artist').count().shape[0])
solo_artisti.fillna(0, inplace=True)
solo_artisti=solo_artisti[solo_artisti['nome_x']==0]
solo_artisti = solo_artisti.rename(columns={'nome_y':'nome'})
solo_artisti.drop(columns=['edizione','nome_x'], inplace=True)
solo_artisti= solo_artisti.drop_duplicates(subset='nome')
print('POST RIMOZIONE MISSING VALUES')
print(solo_artisti.shape)
print('solo_artisti:',solo_artisti.groupby('id_artist').count().shape[0])
solo_artisti = solo_artisti.reset_index(drop=True)
solo_artisti.head(3)
```

```python
print('solo_cantanti',solo_cantanti.groupby('id_artist').count().shape[0])
print('num_solo_artisti:',solo_artisti.groupby('id_artist').count().shape[0])
print('num_artisti_autori:', artisti_autori.groupby('id_artist').count().shape[0])
```

## solo gruppi

```python
solo_gruppi = pd.read_csv('/content/drive/MyDrive/Data_man_Project/Data_Man_project/Data/SOLO_GRUPPI_artisti_final.csv', quotechar="'")
print(solo_gruppi.shape)
print('num_gruppi:',solo_gruppi.groupby('nome').count().shape[0])
solo_gruppi.drop(columns='edizione', inplace=True)
solo_gruppi = solo_gruppi.drop_duplicates().reset_index(drop=True)
print('POST RIMOZIONE MISSING VALUES')
print(solo_gruppi.shape)
print('num_gruppi:',solo_gruppi.groupby('nome').count().shape[0])
solo_gruppi.nome=solo_gruppi.nome.apply(lambda x: str.lower(x))
solo_gruppi.head(5)
```

## generi

```python
generi = pd.read_csv('/content/drive/MyDrive/Data_man_Project/Data_Man_project/Data/generi.csv')
print(generi.shape)
generi=generi.drop_duplicates().reset_index(drop=True)
print('POST RIMOZIONE DUPLICATI')
print(generi.shape)
print('missing values')
print(generi.isna().sum())
## rimozione missing values
generi.dropna(subset=['genres'], inplace=True)
print('POST RIMOZIONE MISSING VALUES')
print(generi.shape)
## convertiamo colonna 'genres' in lista per poi esploderla
generi.genres = generi.genres.apply(lambda x: x.split(', '))
generi = generi.explode('genres').reset_index(drop=True)
print('POST ESPLOSIONE')
print(generi.shape)
print('generi_unici:',generi.groupby('genres').count().shape[0])
print('artisti_unici:',generi.groupby('id_artist').count().shape[0])
generi.head(3)
```

```python
## creazione nodi generi singolo
nodi_gen = pd.DataFrame(set(generi.genres), columns=['genere'])
print(nodi_gen.shape)
nodi_gen.head(3)
```

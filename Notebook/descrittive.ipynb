```python
autori2 = autori.groupby('Autori').count().reset_index()
autori2
```

```python
num = list(autori2.id_track.unique())
```

```python
num
```

```python
prova=pd.DataFrame(columns=['canzoni scritte', 'percentuale'])
num_canzoni=[]
percentuali=[]
tot = autori2.shape[0]
print(tot)
tot2=0
for i in num:
    num_canzoni.append(i)
    a = autori2[autori2['id_track']==i].shape[0]
    
    b = (a/tot*100)
    percentuali.append(b)
    print('percentuale',i,'canzoni scritte',b,'%')
    tot2=tot2+b
    print('cumulativo:', tot2)
prova['canzoni scritte']=num_canzoni
prova['percentuale']=percentuali
prova
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data_man_Project/Data_Man_project/Data/pivot_artisti_final.csv')
df = df.groupby('id_track').agg({'id_artist':list}).reset_index()
df['conto']=df.id_artist.apply(lambda x: len(x))
df.head(2)
```

```python
df = df[df['conto']>1]
df['id_artist2'] = df['id_artist']
df = df.explode('id_artist')
df = df.explode('id_artist2')
df=df.reset_index(drop=True)
df
```

```python
same_person = []
for i, x in enumerate (df.iterrows()):
    art = df.loc[i,'id_artist']
    art1 = df.loc[i,'id_artist2']
    if art==art1:
        same_person.append('stesso')
    else:
        same_person.append('diverso')
df['persona'] = same_person      
df.head(5)
```

```python
df = df[df['persona']!='stesso'].reset_index(drop=True)
df.head()
```

```python
df['inverso']='x'
```

```python
for i, x in enumerate(df.iterrows()):
    for a, b in enumerate(df.id_track):
        
        if df.loc[i, 'id_track'] == b and i!=a:
            
            if df.loc[i, 'id_artist'] == df.loc[a, 'id_artist2'] and df.loc[i, 'id_artist2'] == df.loc[a, 'id_artist']:
                if i > a:
                    df.loc[i, 'inverso']='non va bene'
                else:
                    df.loc[i, 'inverso']='va bene'
                
df.head(5)
```

```python
df[df['conto']==3].head(6)
```

```python
df = df[df['inverso']!='non va bene'].reset_index(drop=True)
df.head(5)
```

```python
df.isna().sum()
```

```python
df = df.merge(tracce[['id_track', 'edizione', 'titolo']], on='id_track', how='left')
df.head(5)
```

```python
art = []
for artist in df.id_artist:
    art.append(artist)
    
for artist2 in df.id_artist2:
    art.append(artist2)
print(len(art))
```

```python
print(df.shape)
print('canzoni:',df.groupby('id_track').count().shape[0])
print('artisti:',len(set(art)))
```

## MEMBRI GRUPPI

```python
gruppi = pd.read_csv('/content/drive/MyDrive/Data_man_Project/Data_Man_project/Data/SOLO_GRUPPI_artisti_final.csv')
print(gruppi.nunique())
gruppi.nome=gruppi.nome.apply(lambda x: x.lower())
gruppi
```

```python
lista_gruppi = []
for x in gruppi.nome:
  if x not in lista_gruppi:
    lista_gruppi.append(x)
lista_gruppi
lista_gruppi.remove('frenetik&orang3')
```

```python
membri = pd.DataFrame(columns={'gruppo', 'membri'})
membri
```

```python
gru=[]
mem=[]
for i, d in enumerate(lista_gruppi):
    wiki = wikipedia.page(d+' gruppo').content
    
    if '== Formazione attuale' in wiki:
        wiki = wiki.split('== Formazione attuale')[1].split('== E')[0]
        gru.append(d)
        mem.append(wiki)
        print(d, wiki)
    elif '== Formazione' in wiki and 'Ex componenti' in wiki:
        wiki = wiki.split('== Formazione')[1].split('== D')[0]
        print(d, wiki)
        gru.append(d)
        mem.append(wiki)
    elif '== Formazione' in wiki and 'Ex componenti' not in wiki:
        wiki = wiki.split('== Formazione ==')[1].split('== D')[0]
        print(d, wiki)
        gru.append(d)
        mem.append(wiki)
    elif 'composto da' in wiki:
        wiki = wiki.split('composto da')[1].split('.')[0]
        #print(d, wiki)
        gru.append(d)
        mem.append(wiki)
    elif 'formato da' in wiki:
        wiki = wiki.split('formato da')[1].split('.')[0]
        #print(d, wiki)
        gru.append(d)
        mem.append(wiki)
```

```python
membri['gruppo'] = gru
membri['membri'] = mem
membri
```

```python
membri.membri = membri.membri.apply(lambda x: x.replace('=',''))
#membri.membri = membri.membri.apply(lambda x: x.replace('\n',''))
#membri.membri = membri.membri.apply(lambda x: .split('Formazione')[1])
```

```python
membri.membri=membri.membri.apply(lambda x: x.split('\n'))
membri = membri.explode('membri').reset_index(drop=True)
membri
```

```python
membri2 = membri[membri['membri']!=''].reset_index(drop=True)
membri2 = membri2[membri2['membri']!=' '].reset_index(drop=True)
membri2 = membri2[membri2['membri']!='='].reset_index(drop=True)
membri2 = membri2[membri2['membri']!=' =='].reset_index(drop=True)
membri2 = membri2[membri2['membri']!=' 24 giugno 1993)'].reset_index(drop=True)
membri2 = membri2[membri2['membri']!= ' 4 ottobre 1994)'].reset_index(drop=True)
membri2 = membri2[membri2['membri']!=' 11 febbraio 1995)'].reset_index(drop=True)
membri2
```

```python
membri2.membri = membri2.membri.apply(lambda x: str(x))
membri2.membri = membri2.membri.apply(lambda x: x.strip())
membri2.membri = membri2.membri.apply(lambda x: x.replace('-', '–'))
```

```python
for i, x in enumerate(membri2.iterrows()):
    print('GRUPPO',i, membri2.loc[i, 'gruppo'])
    print(membri2.loc[i, 'membri'])
    print(type(membri2.loc[i, 'membri']))
    if '(' in membri2.loc[i, 'membri'] and membri2.loc[i, 'gruppo'] != 'pinguini tattici nucleari':
        membri2.loc[i, 'membri'] = membri2.loc[i, 'membri'].split('(')[0]
        print('MODIFIED: ',membri2.loc[i, 'membri'])
    if '–' not in membri2.loc[i,'membri']:
        membri2.loc[i, 'membri'] = membri2.loc[i, 'membri'] + ' –manca strumento'
        
```

```python
membri2['nome'] = membri2['membri'].apply(lambda x: x.split('–')[0])
membri2['strumento'] = membri2['membri'].apply(lambda x: x.split('–')[1])
membri2
```

```python
membri2.groupby('gruppo').count()
```

```python
## metto a posto riga con sir bob cornelius rifo, in quanto in autori è preente bob cornelius rifo
membri2.loc[20,'nome'] =membri2.loc[20,'nome'].split('Sir')[1].split(',')[0]
## togliamo soprannomi tra ""
membri2.nome  = membri2.nome.apply(lambda x: re.sub('"(.*?)"','', x))
## togliamo caratteri tra ()
membri2.nome  = membri2.nome.apply(lambda x: re.sub('\((.*?)\)','', x))
membri2.nome  = membri2.nome.apply(lambda x: x.split('(')[0])
membri2.nome  = membri2.nome.apply(lambda x: x.split(' alias ')[0])
membri2.nome  = membri2.nome.apply(lambda x: x.strip())
## eliminare eventuali doppi spazi
membri2.nome  = membri2.nome.apply(lambda x: " ".join(y for y in x.split() if y not in [' ']))
## tutto in minuscolo
membri2.nome  = membri2.nome.apply(lambda x: x.lower())
```
